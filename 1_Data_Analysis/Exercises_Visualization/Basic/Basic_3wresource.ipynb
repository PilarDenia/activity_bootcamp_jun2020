1.- Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd

%matplotlib inline
```

```python
x = np.linspace(0,48,25)
y = x*3
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw a line')
plt.show()
```

```python
# 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.figure()
plt.plot(x1,y1, label='Line 1', c='b')
plt.plot(x2,y2, label='Line 2', c='g')
plt.legend()
plt.title('Two o more lines in the same plot with suitable legends!')
plt.xlabel('x-label')
plt.ylabel('y-axis');
```

```python
# 6. Write a Python program to plot two or more lines with legends, different widths and colors.
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.figure()
plt.plot(x1,y1, label='Line 1 - width=2', c='b',linewidth=2, linestyle='-')
plt.plot(x2,y2, label='Line 2 - width=5', c='r',linewidth=5, linestyle='-')
plt.legend()
plt.title('Two o more lines in the same plot with suitable legends!')
plt.xlabel('x-label')
plt.ylabel('y-axis');
```

```python
#8. Write a Python program to plot two or more lines and set the line markers. 
x = np.arange(1,8,1)
print(x)
y1 = np.linspace(2,6,4)
y2 = ([3,6,3])
y = np.concatenate([y1,y2])
y
```

```python
plt.plot(x,y,c='r', linestyle=':', linewidth=2, markerfacecolor='blue', marker='o',markersize=12)
plt.title('Display marker')
plt.xlim(1,8)
plt.ylim(1,8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# 14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks.
f, ax = plt.subplots()
plt.plot(df['Close'],marker='o', color='r')
x_min = pd.to_datetime(df.index.min())
x_max = pd.to_datetime(df.index.max())

plt.minorticks_on()

ax.grid(which='major', linestyle='-', c='r', linewidth=0.6)
ax.grid(which='minor', linestyle=':', color='grey')
plt.title('Closing stock value of Alphabet Inc')
plt.ylabel('Closing value')
plt.xlabel('Date')
plt.show()
```

```python
# 15. Write a Python program to create multiple plots. 
grid = plt.GridSpec(2, 3, wspace=0.15, hspace=0.25)
```

```python
plt.subplot(grid[0,:])
plt.tick_params(bottom=False, labelbottom=False, left=False, labelleft=False)
plt.subplot(grid[1,0:1])
plt.tick_params(bottom=False, labelbottom=False, left=False, labelleft=False)
plt.subplot(grid[1,1:2])
plt.tick_params(bottom=False, labelbottom=False, left=False, labelleft=False)
plt.subplot(grid[1,2:3])
plt.tick_params(bottom=False, labelbottom=False, left=False, labelleft=False)
```
