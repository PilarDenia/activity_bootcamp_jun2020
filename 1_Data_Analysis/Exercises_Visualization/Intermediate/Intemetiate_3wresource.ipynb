```python
# 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
dict = {'Languages': languages,'Popularity':popularity}
df = pd.DataFrame(dict)
df
```

```python
f, ax = plt.subplots()
ax.grid(which='major', color='r', linewidth=0.5, alpha=0.5)
ax.grid(which='minor',linestyle=':', color='grey')
ax.minorticks_on()
plt.bar(df.Languages, df.Popularity, color='blue')
plt.title('Popularity of Programing Language\n Wordwide, Oct 2017 compared to a year ago')
ax.set(ylabel='Popularity', xlabel='Language',ylim=(0,25));
```

```python
# 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
```

```python
df = pd.DataFrame({'mean_men':[22, 30, 35, 35, 26], 'mean_women':[25, 32, 30, 35, 29]}, index=['G1','G2','G3','G4','G5'])
df
```

```python
f, ax = plt.subplots()

ind = np.arange(len(df.mean_women))  # the x locations for the groups
width = 0.40  # the width of the bars

ax.bar(ind - width/2,df.mean_men, width=width, color='g', label='men')
ax.bar(ind + width/2, df.mean_women, width=width, color='red', label='women')

#Set number ticks x_axis
ax.set_xticks(ind)
#Set name ticks x_axis
ax.set_xticklabels(df.index)

ax.legend()
ax.set_title('Scores by person')
ax.set_ylabel('Score')
ax.set_xlabel('Person')
ax.set_ylim(0,40)
```

```python
# 14. Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)
```

```python
df = pd.DataFrame({'mean_men':[22, 30, 35, 35, 26], 'mean_women':[25, 32, 30, 35, 29]}, index=['G1','G2','G3','G4','G5'])
df
```

```python
df['std_men'] = [4, 3, 4, 1, 5]
df['std_women'] = [3, 5, 2, 3, 3]
df
```

```python


width = 0.45       # the width of the bars: can also be len(x) sequence

plt.bar(df.index, df.mean_men, width= width, yerr=df.std_men, color='red', label = 'Men', bottom= df.mean_women)
plt.bar(df.index, df.mean_women, width = width, yerr=df.std_women, color='green', label='Women')

plt.ylabel('Scores')
plt.title('Scores by group and gender')

plt.grid(which='major', color='g', linewidth=0.3)
plt.grid(which='minor', color='grey', linestyle=':', alpha=0.3 )
plt.minorticks_on()

ind = np.arange(len(df.index))    # the x locations for the groups
plt.xticks(ind, df.index)
plt.yticks(np.arange(0, 81, 10))
plt.legend();
```
