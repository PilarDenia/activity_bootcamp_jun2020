```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
eje = np.linspace(0,5,100)

#Base del tablero donde pintar, en este caso es de base cuadrada
X,Y = np.meshgrid(eje,eje)

def f(x,y):
    return 3*x + 2*y

ax= plt.axes(projection='3d')

ax.plot_surface(X,Y, f(X,Y))
```

```python
eje_x = np.linspace(0,5,20)
eje_y = np.linspace(0,10,20)
z = 4

#Base del tablero donde pintar, en este caso es de base cuadrada
X,Y = np.meshgrid(eje_x,eje_y)

def f(x,y):
    return ((x**3) + (y*2)) /-z

ax= plt.axes(projection='3d')

ax.plot_surface(X,Y, f(X,Y))
```

```python
eje_x = np.linspace(0,5,25)
eje_y = np.linspace(0,10,15)
z = 10

#Base del tablero donde pintar, en este caso es de base cuadrada
X,Y = np.meshgrid(eje_x,eje_y)

def f(x,y):
    return ((x**3) + (y*2)) /-z

ax1= plt.axes(projection='3d')

ax1.scatter(X, Y, f(X,Y), c='g', marker='o')
```

```python
# Vamos a dibujar una superficie que no sea plana
```

```python
eje = np.linspace(0,5,25)
 
X,Y = np.meshgrid(eje,eje)

def f(x,y):
    return 2*x**2 + 3 * y **2 -0.2

ax = plt.axes(projection = '3d')

ax.plot_wireframe(X,Y,f(X,Y), color='r');
```
