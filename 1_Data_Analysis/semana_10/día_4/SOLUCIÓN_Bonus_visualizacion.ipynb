```python
import numpy as np
import matplotlib.pyplot as plt

mi_np_array = np.array([2,3,4,5])
```

```python
# suma 2, 3 y 4 al array y píntalo
```

```python
suma = [2,3,4]
len(suma)
```

```python
eje_x = range(4)

for x in range(len(suma)):
    plt.plot(eje_x, mi_np_array + suma[x])
```

```python
# ahora en un solo paso con np.newaxis
```

```python
mi_np_array.shape # una sola dimensión
```

```python
mi_np_array
```

```python
mi_np_array_2D = mi_np_array[:, np.newaxis]
mi_np_array_2D
```

```python
mi_np_array_2D.shape # 2 dimensiones, filas y columnas
```

```python
suma_por_filas = mi_np_array_2D + suma
suma_por_filas 
```

```python
plt.plot(eje_x, suma_por_filas)
```

Ahora ya sabemos cómo pasar de 1D a 2D en un array con np.newaxis y que nos podemos ahorrar varios plot

```python
# Vamos a dibujar un plano inclinado
```

```python
eje = np.linspace(0,5,100)

X, Y = np.meshgrid(eje, eje)

def f(x,y):
    return 3*x + 2*y

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,f(X,Y));
```

```python
# Vamos a dibujar una superficie que no sea plana
```

```python
eje = np.linspace(0,5,100)

X, Y = np.meshgrid(eje, eje)

def f(x,y):
    return 2 * x**2 + 3 * y**2 - 0.2

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,f(X,Y));
```
