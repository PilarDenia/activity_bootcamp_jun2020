### Splitting Data into Training and Test Sets (Digits Dataset)

```python
from sklearn.model_selection import train_test_split

X = digits.data
y = digits.target
print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', multi_class='auto')
```

```python
model.fit(X_train,y_train)
```

```python
print('Shape X_train:',X_train.shape)
print('Shape X_test:',X_test.shape)
print('Shape y_train:',y_train.shape)
print('Shape y_test:',y_test.shape)
```

```python
#usamos el método de score para obtener el accuracy del modelo
score = model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
predictions = model.predict(X_test)
c_matrix = confusion_matrix(y_test,predictions)
c_matrix
# Columnas ==> Prediccion  y Filas/Row ==> Realidad
```

```python
import seaborn as sns
sns.heatmap(c_matrix,annot=True, cmap='Greens')
# Columnas ==> Prediccion  y Filas/Row ==> Realidad
plt.xlabel('Predicted label')
plt.ylabel('Actual label')
all_sample_title = ('Accuracy: '+str(round((score * 100),2)) +'%')
plt.title(all_sample_title + '\n',fontsize=15);
```

```python
row_sums = c_matrix.sum(axis=1, keepdims = True) #Comprobamos si las clases están balanceadas ¿?
row_sums
```

```python
norm_conf_mx = c_matrix / row_sums
np.fill_diagonal(norm_conf_mx, 0) # Relleno la diagonal con '0' 

#Quiero visualizar donde se falla 
plt.figure(figsize=(8,8))
sns.heatmap(norm_conf_mx, annot=True, linewidths=.5, square=True, cmap='Greens')
plt.xlabel('Predicted label')
plt.ylabel('Actual label')
all_sample_title = (f'Accuracy: {str(round((score * 100),2))} % \n')
plt.title(all_sample_title + '\n', fontsize=15);
```
