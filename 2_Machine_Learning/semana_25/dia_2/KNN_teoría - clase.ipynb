## Algoritmo KNN

El algoritmo KNN (K Nearest Neighbors) trata de buscar los K puntos más cercanos a un punto concreto para poder inferir su valor.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

from sklearn.neighbors import KNeighborsClassifier
```

```python
%matplotlib inline
```

Tenemos la altura y el peso de cinco personas, junto con su respectivo género (Hombre o mujer).

```python
data = {'Peso': [50, 80, 90, 45, 60],
        'Altura': [1.48, 1.82, 1.85, 1.55, 1.60],
        'Genero': ['m', 'h', 'h', 'm', 'm']}

df = pd.DataFrame(data)
df
```

```python
ax = plt.axes()

ax.scatter(df.loc[df['Genero'] == 'h', 'Peso'], 
                  df.loc[df['Genero'] == 'h', 'Altura'],
                  color = 'r',
                  label = 'Hombre')

ax.scatter(df.loc[df['Genero'] == 'm', 'Peso'],
           df.loc[df['Genero'] == 'm', 'Altura'],
           c="b",
           label="Mujer")

plt.xlabel('Pesos')
plt.ylabel('Altura')
ax.legend();
```

Si nos dieran datos de una sexta persona, pero no nos dieran su género, ¿Podríamos adivinarlo de alguna forma?

```python
pto_nuevo = {'Peso':[70],
             'Altura':[1.82]}

pto_nuevo = pd.DataFrame(pto_nuevo)


ax = plt.axes()

ax.scatter(df.loc[df['Genero'] == 'h', 'Peso'], 
                  df.loc[df['Genero'] == 'h', 'Altura'],
                  color = 'r',
                  label = 'Hombre')

ax.scatter(df.loc[df['Genero'] == 'm', 'Peso'],
           df.loc[df['Genero'] == 'm', 'Altura'],
           c="b",
           label="Mujer")

ax.scatter(pto_nuevo['Peso'],
           pto_nuevo['Altura'],
           color = 'k',
           label = 'new_person')

plt.xlabel('Pesos')
plt.ylabel('Altura')
ax.legend();
```

```python
knn = KNeighborsClassifier(n_neighbors=3)

X = df[['Peso','Altura']]
y = df['Genero']

knn.fit(X,y)
```

```python
print(y.shape)
```

```python
prediccion = knn.predict(pto_nuevo)
print(prediccion)
```
